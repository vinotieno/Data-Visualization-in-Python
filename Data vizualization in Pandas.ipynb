```python
df = pd.read_csv(r'C:\Users\ADMIN\Downloads\Visualization in pandas Ice Cream Ratings.csv')

df = df.set_index('Date')

df
```

```python
print(plt.style.available)

plt.style.use('classic')
```

#### Plot() in pandas

```python
df.plot()
```

#### Line plot 

```python
df.plot(kind = 'line')
```

```python
df.plot(kind = 'line', subplots = True)
```

```python
df.plot(kind = 'line', title = 'Ice Cream Rating')
```

```python
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

#### Bar plot

```python
df.plot(kind = 'bar')
```

```python
df.plot(kind = 'bar', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot(kind = 'bar', stacked = True)
```

```python
df.plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df['Flavor Rating'].plot(kind = 'bar', stacked = True, title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
df.plot.barh(stacked = True)
```

### Scatter Plot
